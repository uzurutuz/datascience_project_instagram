## Socioal Characteristics Data


```python
def get_median_categorical(size, df):
    
    df_cumulative = df
    
    median_size = size / 2

    for i in range(1, len(df_cumulative.columns)):
        df_cumulative.iloc[:, i] = df_cumulative.iloc[:, i - 1] + df_cumulative.iloc[:, i]

    cols = df_cumulative.columns
    df_cumulative['median'] = ''
    
    for i, row in df.iterrows():
        vals = np.array(row[:-1].values)
        indx = np.where(vals >= median_size[i])[0][0]
        df_cumulative.loc[i, 'median'] = cols[indx]
        
    return df_cumulative['median']
    pass

df_social = pd.read_csv("Data/social_characteristics_PIT/social_characteristics_PIT.csv")
df_social.columns =['zipcode', 'avg_household_size', 'pop_over_25', 'less_9th_grade', '9th_to_12th_grade', 'hihgh_school_grad', 'some_college_no_degree', 'associate_degree', 'bachelor_degree', 'graduate_or_professional_degree']
df_social = df_social.set_index('zipcode')
df_social

df_social['median_education'] = get_median_categorical(df_social['pop_over_25'], df_social[['less_9th_grade', '9th_to_12th_grade', 'hihgh_school_grad', 'some_college_no_degree', 'associate_degree', 'bachelor_degree', 'graduate_or_professional_degree']])
df_social
```
